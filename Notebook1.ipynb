```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./spends.csv")
```

```python
df['Category'].unique()
```

```python
df['Category'] = df['Category'].replace('Housing and Utilities', 'housing')
```

```python
df['Category'] = df['Category'].replace('Friend Activities', 'friend')
```

```python
df['Category'] = df['Category'].replace('Medical/Dental', 'medical')
```

```python
df['Category'] = df['Category'].str.title()
```

```python
df['Category'].unique()
```

```python
df['Customer ID'].duplicated().sum()
```

```python
# Sort by Customer ID and save it back to df
df1 = df.sort_values(by='Customer ID')
```

```python
df1 = df1.drop(columns=["Item"])
```

```python
df1
```

```python
# Sort by ID first (primary), then by Category (secondary)
df1 = df.sort_values(by=['Customer ID', 'Category'])
```

```python
df1 = df1.drop(columns=['Item'])
```

```python
df1
```

```python
# Group by Customer and Category
df1_aggregated = df1.groupby(['Customer ID', 'Category'], as_index=False).agg({
    'Total Spent': 'sum',           # Add up all the money
    'Quantity': 'sum',              # Add up total items
    'Transaction Date': 'max',      # Keep the most recent transaction date
    'Payment Method': 'first',      # Keep the first payment method used
    'Location': 'first'             # Keep the first location recorded
})

# View the result for CUST_0001 to see the single Fitness row
print(df1_aggregated[df1_aggregated['Customer ID'] == 'CUST_0001'])
```

```python
df_final = df.groupby(['Customer ID', 'Category'], as_index=False).agg({
    'Total Spent': 'sum',            
    'Quantity': 'sum',               
    'Transaction Date': 'max',       
    'Payment Method': 'first',       
    'Location': 'first'              
})

print(f"Original rows: {len(df)}")
print(f"Condensed rows: {len(df_final)}")
print(df_final.head(10))
```

```python
df_final = df_final.sort_values(by=['Customer ID', 'Category'])
```
